# **1. Business Problem & Understanding**

## **1.1 Latar Belakang**

Customer Lifetime Value (CLV) adalah metrik penting bagi perusahaan untuk mengetahui seberapa berharga seorang pelanggan, dan seberapa banyak total pendapatan yang didapatkan perusahaan dari setiap customer selama rentang waktu tertentu. Dengan memprediksi CLV, perusahaan bisa melakukan strategi pemasaran yang lebih akurat, meningkatkan kepuasan pelanggan, dan mendapatkan lebih banyak revenue. Memahami faktor-faktor yang mempengaruhi CLV membantu perusahaan dalam menggunakan sumber daya dengan lebih baik dan menciptakan program loyalitas yang sesuai dengan kebutuhan pelanggan. Untuk memprediksi CLV, perusahaan dapat menggunakan teknik machine learning jenis regresi, yang menganalisis data pelanggan dan informasi terkait lainnya berdasarkan historis pelanggan dan atribut-atribut terkait.

Pada studi kasus ini, perusahaan asuransi berfokus pada mengidentifikasi dan mengakuisisi customer yang memiliki potensi profitabilitas jangka panjang. Di sini, Customer Lifetime Value (CLV) memegang peranan penting sebagai alat untuk menemukan segmen customer bernilai tinggi berdasarkan informasi data yang tersedia, dan menjadi landasan utama dalam penyusunan strategi yang tepat sasaran. Dengan mengetahui nilai CLV secara akurat, perusahaan asuransi dapat meningkatkan kinerja di berbagai aspek, antara lain:
- Targeted Marketing: Perusahaan dapat menargetkan kepada customer dengan potensi CLV tinggi yang dapat memberikan keuntungan besar bagi perusahaan, serta meningkatkan efektivitas marketing campaign.
- Customer Service: Dapat menggunakan informasi CLV untuk meningkatkan customer experience dan retensi pelanggan, sehingga tidak perlu ada biaya yang besar untuk fokus kepada customer dengan CLV yang rendah.
- Strategic & Planning: Data CLV dapat digunakan untuk merumuskan strategi produk dan layanan yang lebih baik, dan alternatif yang dilakukan untuk mencegah atau menemukan tindakan preventif kepada customer yang churn atau akan churn. Oleh karena itu, perusahaan nantinya dapat mengetahui strategi akuisisi customer yang berpotensi memiliki nilai CLV tinggi dan juga mempertahankan (retention) customer yang sudah ada dengan nilai CLV tinggi.
- Aktuaris: Berkontribusi dalam analisis risiko, penetapan premi, dan klaim yang tepat berdasarkan data CLV.

## **1.4 Evaluasi Metrik**

Dalam proses pemodelan machine learning ini, metrik evaluasi yang digunakan adalah R-Squared, RMSE, MAE, dan MAPE. Keempat metrik ini berfungsi untuk menunjukkan seberapa baik model dalam memprediksi sebuah dataset. R-Squared menunjukkan seberapa baik variabel independen menjelaskan perubahan variabel dependen dalam model. RMSE dipilih sebagai salah satu evaluasi untuk optimisasi, karena RMSE memberikan penalti yang lebih besar pada kesalahan atau outlier yang signifikan, akibat pengkuadratan dari error itu sendiri. Dengan demikian, RMSE sangat cocok untuk meminimalisir kesalahan selama proses optimasi.

$$
R^2 = 1 - \frac{\sum\limits_{i=1}^{n} (Y_i - \widehat{Y_i})^2}{\sum\limits_{i=1}^{n} (Y_i - \bar{Y})^2}
$$

$$
\text{R-squared}
$$

$$RMSE =   \sqrt{ \frac{ \sum\nolimits_{i=1}^{n} ( Y_i - \widehat{Y_i})^2 }{n} } $$
$$Root Mean Squared Error$$

MAE dan MAPE digunakan karena lebih mudah untuk diinterpretasikan. Pada MAE, kesalahan dihitung dengan cara mengurangkan nilai prediksi dari nilai aktual secara rata-rata absolut, sedangkan pada MAPE, kesalahan dihitung dengan rata-rata persentase kesalahan antara nilai yang diprediksi dengan nilai aktual. MAPE juga memiliki keuntungan dalam situasi di mana nilai CLV yang tinggi memiliki kesalahan yang besar, tetapi persentase dari data tersebut sebenarnya rendah.
$$MAE =  \frac{\sum\nolimits_{i=1}^{n} |Y_i - \widehat{Y_i}|}{n}$$
$$Mean Absolute Error$$

$$MAPE =  \frac{1}{n}   \sum_{i=1}^{n} | \frac{Y_i - \widehat{Y_i}}{Y_i}|$$
$$Mean Absolute Percentage Error$$

Keterangan:
- $n$ = Jumlah data<br>
- $i$ = variabel i<br>
- $Y_i$ = Data aktual<br>
- $\widehat{Y_i}$ = Data Prediksi

**2. Pengaruh Vehicle Class Terhadap CLV**

```python
# Analyze the average 'Customer Lifetime Value' by 'Vehicle Class'
clv_vehicle_class = df.groupby('Vehicle Class')[['Customer Lifetime Value']].median()
clv_vehicle_class.sort_values(by="Customer Lifetime Value", ascending=False)
```

```python
# Uji Hipotesis Menggunakan Kruskal Walls

# Hipotesis Nol (H0): Tidak ada perbedaan signifikan dalam nilai CLV antara kelompok-kelompok yang berbeda dari Vehicle Class.
# Hipotesis Alternatif (H1): Ada perbedaan signifikan dalam nilai CLV antara setidaknya satu kelompok dari Vehicle Class.

stat, pvalue = kruskal(
    df[df['Vehicle Class']=='Four-Door Car']['Customer Lifetime Value'],
    df[df['Vehicle Class']=='Luxury Car']['Customer Lifetime Value'],
    df[df['Vehicle Class']=='Luxury SUV']['Customer Lifetime Value'],
    df[df['Vehicle Class']=='SUV']['Customer Lifetime Value'],
    df[df['Vehicle Class']=='Sports Car']['Customer Lifetime Value'],
    df[df['Vehicle Class']=='Two-Door Car']['Customer Lifetime Value']
)

# conclusion

if pvalue <= 0.05:
    print(f'''
    pvalue = {pvalue}. pvalue < 0.05. Tolak H0. 
    Ada perbedaan signifikan dalam nilai CLV antara setidaknya satu kelompok dari Vehicle Class.
    ''')
else:
    print(f'''
    pvalue = {pvalue}. pvalue > 0.05. Terima H0. 
    Tidak ada perbedaan signifikan dalam nilai CLV antara kelompok-kelompok yang berbeda dari Vehicle Class.
    ''')
```

## **3.3 Pre-processing**

### **3.3.1 Scaling & Encoding**

**Encoding:**
- **OneHot Encoder :** `Renew Offer Type`, `Martial Status`

    OneHot Encoding digunakan untuk mengubah variabel kategorikal yang tidak memiliki urutan menjadi format numerik yang dapat digunakan oleh model machine learning. Nantinya, setiap kategori akan menjadi kolom baru dengan nilai 1 dan 0. Pada kasus ini, `Renew Offer Type` dan `Martial Status` tidak memiliki urutan yang jelas, sehingga OneHot Encoding adalah pilihan yang tepat.

- **Ordinal Encoder :** `Vehicle Class`, `Coverage`, `Education`

    Ordinal Encoding digunakan untuk variabel kategorikal yang memiliki urutan atau ranking. Nantinya, setiap kategori akan menjadi angka numerik 1 sampai n sesuai dengan jumlah data unique pada feature. Pada kasus ini, `Vehicle Class`, `Coverage`, dan `Education` memiliki urutan berdasarkan tingkat kendaraan, tingkat perlindungan yang dipilih, dan tingkat pendidikan.

- **Binary Encoder :** `Employment Status`

    Binary Encoding adalah metode yang efisien untuk mengonversi variabel kategorikal dengan banyak kategori menjadi format biner.

**Scaling:**
- **Robust Scaling :** `Number of Policies`, `Monthly Premium Auto`, `Total Claim Amount`, `Income`, `Vehicle Class`, `Coverage`, `Education`

    Robust Scaling digunakan untuk mengatasi outlier dalam data, mengingat data yang ada terdapat banyak outlier. Dengan menggunakan median dan interquartile range (IQR), Robust Scaling memastikan bahwa fitur-fitur tersebut dinormalisasi tanpa terpengaruh oleh nilai ekstrem.

```python
ordinal_mapping = [{'col': 'Vehicle Class',
                    'mapping': {'Luxury SUV': 6, 'Luxury Car': 5, 'Sports Car': 4, 'SUV':3, 'Four-Door Car': 2, 'Two-Door Car':1}},
                    {'col': 'Coverage',
                    'mapping': {"Basic" : 1,"Extended" : 2, "Premium" : 3}},
                    {'col': 'Education',
                    'mapping': {"High School or Below" : 1,"College" : 2, "Bachelor" : 3, "Master" : 4, "Doctor" : 5}}]
```

```python
pipe_ordinal_robust = Pipeline([
    ("ordinal", OrdinalEncoder(mapping=ordinal_mapping)),
    ("robust", RobustScaler())
])
```

```python
transformer = ColumnTransformer([
    ("one_hot_encoder", OneHotEncoder(),["Renew Offer Type","Marital Status","EmploymentStatus"]),
    ("robust_scaler", RobustScaler(),["Number of Policies", "Monthly Premium Auto", "Total Claim Amount", "Income"]), 
    ("pipe_ordinal_robust", pipe_ordinal_robust,["Vehicle Class", "Coverage", "Education"])
], remainder="passthrough")
# Robust scaler, mengingat data tidak terdistribusi normal dan untuk tackle outlier
# Encoder, untuk mengubah data kategori menjadi numerik

transformer
```

### **3.3.2 Cross Validation**

Di tahap ini, kita akan mencari algoritma model benchmark untuk mengevaluasi performa berbagai model regresi. Mengingat bahwa target dalam dataset ini adalah variabel numerik kontinu, kita akan fokus pada algoritma jenis regresor.

Kita akan mencoba beberapa model, seperti Linear Regression, Ridge, Lasso, Decision Tree Regressor, Random Forest Regressor, Gradient Boosting Regressor, XGBoost Regressor, dan KNeighbors Regressor. Untuk memastikan hasil yang lebih akurat dan menghindari overfitting, kita juga akan melakukan cross-validation dengan 5 fold yang berarti nantinya akan ada 4 bagian yang akan ditrain dan 1 bagian menjadi validasi.

```python
# Menentukan algoritma model benchmark
# Untuk menetralkan hasil dari model benchmark, sebisa mungkin tidak perlu setting parameter kecuali random state

linear_regression = LinearRegression()
ridge = Ridge(random_state=0)
lasso = Lasso(random_state=0)
decision_tree = DecisionTreeRegressor(random_state=0)
random_forest = RandomForestRegressor(random_state=0)
gradient_boost = GradientBoostingRegressor(random_state=0)
xgb = XGBRegressor()
knn = KNeighborsRegressor()
```

```python
# model
list_model = [linear_regression, ridge, lasso, knn, decision_tree,random_forest,gradient_boost,xgb]

# list scorer
list_scorer = ['neg_root_mean_squared_error', 
 'neg_mean_absolute_error',
 'neg_mean_absolute_percentage_error',
 'r2']

# list kosong untuk menampung score
list_mean_rmse = []
list_std_rmse = []

list_mean_mae = []
list_std_mae = []
    
list_mean_mape = []
list_std_mape = []

list_mean_r2= []
list_std_r2 = []

list_kosong = []

# looping
for i in list_model:
    # pipeline
    pipe_prep_model = Pipeline([
        ('preprocessing', transformer),
        ('modeling', i)
    ])

    
    # cross validate
    # untuk melakukan cross validation menggunakan beberapa metrics evaluation
    crossval = cross_validate(estimator = pipe_prep_model,
                    X = X_train,
                    y = y_train,
                    cv = 5,
                    n_jobs = -1,
                    scoring = list_scorer)
    
    # rmse
    list_kosong.append(crossval)
    
    list_mean_rmse.append(crossval['test_neg_root_mean_squared_error'].mean())
    list_std_rmse.append(crossval['test_neg_root_mean_squared_error'].std())
    
    # mae
    list_mean_mae.append(crossval['test_neg_mean_absolute_error'].mean())
    list_std_mae.append(crossval['test_neg_mean_absolute_error'].std())
    
    # mape
    list_mean_mape.append(crossval['test_neg_mean_absolute_percentage_error'].mean())
    list_std_mape.append(crossval['test_neg_mean_absolute_percentage_error'].std())

    # mape
    list_mean_r2.append(crossval['test_r2'].mean())
    list_std_r2.append(crossval['test_r2'].std())
```

```python
df_score = pd.DataFrame()
df_score["Algo"] = ["Linear Regression", "Ridge", "Lasso", "K-Neighbors Regressor", "Decision Tree Regressor", "Random Forest Regressor", "Gradient Boost Regressor", "XGB Regressor"]

df_score["mean_score_RMSE"] = list_mean_rmse
df_score["std_score_RMSE"] = list_std_rmse

df_score["mean_score_MAE"] = list_mean_mae
df_score["std_score_MAE"] = list_std_mae

df_score["mean_score_MAPE"] = list_mean_mape
df_score["std_score_MAPE"] = list_std_mape

df_score["mean_score_R2"] = list_mean_r2
df_score["std_score_R2"] = list_std_r2

df_score
```

## **3.4 Hyperparameter Tuning**

### **3.4.1 Hyperparameter Tuning Model Gradient Boost**

Pada bagian ini, kita akan melakukan tuning hyperparameter untuk model Gradient Boosting. Proses ini melibatkan penentuan nilai optimal untuk beberapa parameter penting, seperti kedalaman pohon, learning rate, jumlah pohon, minimum sample split dan minimum sample leaf.

**1. Gradient Boost**

Gradient Boost adalah teknik ensemble learning berbasis decision tree yang digunakan dalam supervised learning yang bekerja dengan membangun model secara bertahap, di mana setiap model baru berusaha memperbaiki kesalahan model sebelumnya. Berikut adalah contoh visualisasinya.

![Gradient Boost Explanation (2).png](<attachment:Gradient Boost Explanation (2).png>)

- Iterasi 1 (Tree 1 - Pohon Pertama)
    Model pertama mencoba membuat prediksi awal terhadap data. Namun, karena ini hanya satu pohon keputusan, prediksi masih kasar dan tidak terlalu akurat.
    
    Residual/kesalahan → Ditampilkan sebagai garis biru putus-putus, menunjukkan seberapa jauh prediksi dari nilai sebenarnya.

- Iterasi 2 (Menambahkan Pohon Kedua - Tree 1 + Tree 2)
    Pohon kedua dibuat untuk memprediksi residuals dari Tree 1. Residual ini ditambahkan ke prediksi awal, sehingga model menjadi lebih baik, namun masih ada kesalahan terutama pada rentang X = 0,5 sampai X = 2.

- Iterasi 3 (Menambahkan Pohon Ketiga - Tree 1 + Tree 2 + Tree 3)
    Pohon ketiga kembali mempelajari residuals dari model sebelumnya dan mencoba memperbaikinya. Hasilnya, residual semakin kecil jika dibandingkan dengan 2 tahap sebelumnya, yang berarti model semakin akurat dalam menangkap pola pada data.

- Final Prediction
    Setelah 3 pohon, model akhir merupakan gabungan dari semua pohon sebelumnya. Hasil akhirnya hampir sesuai dengan data asli, karena setiap pohon telah membantu mengurangi kesalahan secara bertahap. Residual juga semakin kecil yang menunjukan model semakin baik.

**2. Parameter**

- Max Depth: Menentukan kedalaman maksimum dari setiap decision tree. Semakin dalam pohon, semakin kompleks modelnya, tetapi ini juga dapat menyebabkan overfitting. Tuning parameter ini membantu menemukan kedalaman yang tepat.
- Learning Rate: Mengontrol seberapa besar kontribusi setiap pohon baru terhadap model keseluruhan. Learning rate yang lebih kecil dapat meningkatkan akurasi tetapi memerlukan lebih banyak pohon (n_estimators) untuk mencapai hasil yang baik.
- N Estimators: Jumlah pohon yang akan dibangun dalam model. Semakin banyak pohon, semakin baik model dapat belajar, tetapi ini juga dapat meningkatkan waktu running dan risiko overfitting.
- Min Sample Leaf: Menentukan jumlah minimum sampel yang harus ada di setiap leaf. Parameter ini membantu mengontrol seberapa kompleks suatu model dan mencegah overfitting dengan memastikan bahwa setiap leaf memiliki cukup data.
- Min Sample Split: Menentukan jumlah minimum sampel yang diperlukan untuk membagi node. Ini juga berfungsi untuk mengurangi kompleksitas model dan mencegah overfitting.

**3. Randomized Search**

Randomized Search adalah metode untuk melakukan pencarian hyperparameter yang lebih efisien dibandingkan dengan Grid Search. Jika Grid Search mencoba semua kombinasi parameter, maka Randomized Search akan memilih secara acak kombinasi dari ruang parameter yang ditentukan. Ini memungkinkan eksplorasi yang lebih luas dalam waktu yang lebih singkat.

```python
# Kedalaman pohon
max_depth = list(np.arange(1, 11))

# Learning rate
learning_rate = [0.01, 0.05, 0.1, 0.15, 0.2]

# Jumlah pohon
n_estimators = list(np.arange(50, 301, 50))

# Min sample split
min_sample_split = randint(2, 51)

# Min sample leaf
min_sample_leaf = randint(2, 51)

hyperparam  = {
    'model__max_depth': max_depth, 
    'model__learning_rate': learning_rate,
    'model__n_estimators': n_estimators,
    'model__min_samples_leaf': min_sample_leaf,
    'model__min_samples_split': min_sample_split
}

pipe_prep_model = Pipeline([
        ('preprocessing', transformer),
        ('model', gradient_boost)
        ])

# Random tuning
random_search = RandomizedSearchCV(
    estimator=pipe_prep_model,
    cv=5,
    scoring='neg_root_mean_squared_error',
    n_jobs=-1,
    n_iter=5000,
    param_distributions=hyperparam,
    random_state=0
)

random_search
```
